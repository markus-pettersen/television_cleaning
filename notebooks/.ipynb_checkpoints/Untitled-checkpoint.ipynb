```python
from sqlalchemy import create_engine
import numpy as np
import pandas as pd


engine = create_engine("sqlite:///../data/final/tv_database.db")

# Join shows to languages to also examine language distribution
shows_query = """
    SELECT shows.id, title, start_year, language, end_year, status
    FROM shows
    LEFT JOIN languages
    ON shows.language_id = languages.id
"""

shows_df = pd.read_sql_query(shows_query, engine)
```

```python
test_df = pd.read_csv("../test_df.csv")
```

```python
merged_df = shows_df.merge(test_df, on="id", how="left", suffixes=["_orig", "_miss"])
```
